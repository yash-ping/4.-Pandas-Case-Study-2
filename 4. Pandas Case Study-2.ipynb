```python
#calculate
combination = cust_leftjoin.groupby(['Gender','SeriousDlqin2yrs'])

Number_of_customers = combination['ID'].count()
customers = combination['ID'].count()

customers_sum = customers.sum()
Percentage_of_customers = customers / customers_sum *100
```

```python
#present
Average_Revolving_utilization = combination['RevolvingUtilization'].mean()
Standard_deviation = combination['RevolvingUtilization'].std()
Average_income = combination['MonthlyIncome'].mean()
Maximum_income = combination['MonthlyIncome'].max()
Standard_deviation_income = combination['MonthlyIncome'].std()
Average_age = combination['age'].mean()
Average_Dependents = combination['No_of_dependents'].mean()
```

```python
sum_report = pd.concat([Number_of_customers,Percentage_of_customers,Average_Revolving_utilization,Standard_deviation,Average_income,Maximum_income,Standard_deviation_income,Average_age,Average_Dependents],keys = ['Number of customers',' Percentage of customers','Average Revolving utilization','Standard deviation with in the utilization','Average monthly income','Maximum Monthly income','Standard deviation with in the income','Average Age',' Average Dependents'],axis=1)

sum_report
```

```python
summary_report = cust_leftjoin.groupby(cust_leftjoin['Martial_Status'])


Number_of_married_cust = summary_report['ID'].count()

total_married_customers = Number_of_married_cust.sum()

Percentage_of_married = Number_of_married_cust / total_married_customers *100

Final_report = pd.concat([Number_of_married_cust,Percentage_of_married],keys = ['Number_of_married','Percentage_of_married'],axis=1)
```
